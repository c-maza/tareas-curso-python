# Tarea - Módulo 1

Al mezclar una sal de cadmio $\ce{Cd^{2+}}$ con exceso de $\ce{NH3}$, resulta la generación _in situ_ de un complejo inorgánico $\ce{[Cd(NH3)]^{2+}}$ (entre otros equilibrios) gobernado por una constante de formación, $K_f$, según la siguiente ecuación:

$$\ce{Cd^2+ (ac) + NH3 (ac) <--> [Cd(NH3)]^2+ (ac)} \quad K_f = 10^{2.65}$$

Una vez que el complejo se forma, este permanece en equilibrio con sus iones o moléculas constituyentes como muestra la reacción reversa, pero gobernado por una constante de disociación $K_d$, la cuál se puede determinar a partir de $K_f$ como: 

$$K_d = K_f^{-1}$$

Para una concentración incial de $\ce{Cd^{2+}}$ equivalente a $c_0$, determine la concentración en equilibrio del ión $\ce{Cd^{2+}}$ en el equilibrio. Por simplicidad, asumir que el equilibrio dado es el único que se produce en la reacción y que el cambio de volumen producto de la mezcla de reacciones es despreciable.




El equilibrio de la reacción se define por:
$$
    K_f = \frac{\ce{[Cd(NH3)]^{2+}}}{[\ce{Cd^{2+}}][\ce{NH3}]} = K_d^{-1}
$$
En este caso:
$$
    K_d = \frac{[\ce{Cd^{2+}}][\ce{NH3}]}{\ce{[Cd(NH3)]^{2+}}} = K_f^{-1}
$$
En este caso:
$$
    10^{2.65} = \frac{c_0-x}{x^2}
$$
O bien:
$$
    10^{-2.65} = \frac{x^2}{c_0-x}
$$

Se define el siguiente diagrama para el proceso:



```python
#se asigna el valor a Kf y Kd
K_f = 10 ** 2.65
print(K_f)
K_d = K_f ** -1
print(K_d)
#para comprobar que el valor es correcto:
print(10 ** -2.65)
```

```python
#se define la funcion para encontrar la concentración en equilibrio
def eq(K_f, c_0):
    """
    Determina la concentración de los iones en equilibrio, en base a la concentración inicial del complejo formado (c_0)
    y la constante de formación (K_f)
    """
    K_d = K_f ** -1 #se prefirió usar la cte de disociación, siendo esta el inverso de la de formación.
    #se determinan los valores de a, b, c para la ecuacion cuadratica que resulta del despeje  
    #la ecuacion resultante del despeje es x ** 2 + K_d * x - K_d * c_0
    a = 1
    b = K_d
    c = -(K_d * c_0)
    
    #se considera el valor del discriminante d=b^2-4ac
    d = b ** 2 - (4 * a * c)
    #donde hay 2 posibles condiciones segun su valor:
    #si d > 0 o d = 0 se procede al calculo, si d < 0 la ecuacion no tiene soluciones reales y no tiene sentido fisico,
    #por lo que se descarta.
    if d < 0:
        return(""" !! ERROR: La ecuación no tiene valores reales. Es probable que el valor ingresado de
        la concentración inicial o la constante de formación son negativos, lo que no tiene
        sentido químico. Revisar los datos ingresados y volver a intentar. !! """)
    #se calcula la raiz del discriminante y luego se procede a resolver la ecuación cuadrática
    else:
        raiz_d = d ** 0.5
        x_1 = ( -b + raiz_d) / ( 2 * a )
        x_2 = ( -b - raiz_d) / ( 2 * a )
    #se considera unicamente el valor más alto, siendo este el que tiene sentido (c no puede ser negativa)
        c_eq = max(x_1, x_2)
    #devuelve la concentracion en equilibrio
        return(f"La concentración del ión cadmio (II) en el equilibrio es de {c_eq} mol/L")
```

```python
#Para un valor de c_0 bajo
c_0 = 2.0 * 10 ** -10
print(eq(K_f, c_0))
```

```python
#Para un valor de c_0 medio
c_0 = 4.5
print(eq(K_f, c_0))
```

```python
#Para un valor de c_0 alto
c_0 = 3.0 * 10 ** 15
print(eq(K_f, c_0))
```

```python
#otro calculo
c_0 = 5
print(eq(K_f, c_0))
```

```python
#calculo que lleva a error:
c_0 = -4.0 * 10 ** -2
print(eq(K_f, c_0))
```

Al tener un valor de $K_f = 10^{2.65}$, el equlibrio de la reacción está desplazado hacia los reactivos (los iones)

A valores bajos de $c_0$ (aproximadamente desde $10^{-4}$), casi toda la concentración inicial se mantiene en el ión cadmio(II); al ir subiendo estos valores, más de la concentración inicial reacciona con el ligando, pero aún así parte significativa de esta se mantiene en los iones.
